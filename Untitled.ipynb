```python
import requests
from selenium import webdriver
import numpy as np
import pandas as pd

URL = "https://www.tickpick.com/buy-yeat-tickets-newport-music-hall-4-9-22-7pm/5076846/"

r = requests.get(URL)
```

```python
driver = webdriver.Chrome(r"C:\Users\Paul\Documents\chromedriver.exe")
driver.get(URL)

xpath = "//*[@id=\"listingContainer\"]/div[*]"
old_xpath = "//*[@id=\"listingContainer\"]/div[1]"

elements = driver.find_elements_by_xpath(xpath)
```

```python
rows_list = []

for elem in elements:
    dic = {}
    dic["quantity"] = elem.find_element_by_tag_name("select").text.split()[0]
    dic["price"] = elem.find_element_by_tag_name("b").text[1:]
    rows_list.append(dic)
    
df = pd.DataFrame(rows_list)
df.head()
```
